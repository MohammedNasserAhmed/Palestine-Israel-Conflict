```python
from visualization.piechart import PieChart, PieCharts, Choice
from visualization.bars import  StackBar, vbar, CustomizedBar
from visualization.scatter import Bubbles, Config
from utils.data import Loader
import pandas as pd
Project_Path = "E:\MyOnlineCourses\ML_Projects\Palestine-Israel-Conflict"
loader = Loader()
df = loader.read_csv(Project_Path +"\data\ps_il.csv")

data = df.to_dict(orient='records')
title = "Human Cost of Palestine-Israel Conflict From 2000 To April 2024"
signature = "aiNarabic.ai\nM. N. Gaber\nabunasseredu@gmail.com"
box_title = "TOTAL fatalities and injuries\n           2000 - 2023"
group_var = "Year"
columns = [("Palestinians Injuries","Palestinians Fatalities"),("Israelis Injuries","Israelis Fatalities")]
img_lbls = [("Palestinians","Israelis")]
legend_lbls = [("Injuries","Fatalities")]
psimg = Project_Path+"\images\ps_h.png"
ilimg =Project_Path+"\images\il_h.png"
img_paths = [psimg,
                ilimg,
                psimg,
                ilimg]

map_img = Project_Path+"\images\pmap.png"
save_filename = Project_Path+"\outputs\customizedbar.png"
cdp = CustomizedBar(data=data, 
                    title=title, 
                    box_title=box_title, 
                    gv=group_var, 
                    cols=columns, 
                    img_lbls=img_lbls, 
                    lgd_lbls=legend_lbls, 
                    img_paths=img_paths,
                    map_img=map_img, 
                    signature=signature)
cdp.show_plot(save_filename=Project_Path+"\outputs\customized-bar.png")
```

```python
data = loader.read_excel(Project_Path +"\data\ps_il.xlsx", header=0, usecols=[0,1,2,3,4])
save_filename = Project_Path+"\outputs\Bubbles.html"
title = "Human Cost of Palestine-Israel Conflict<br>2000 To Oct-2023"
config=Config()
scatterplot = Bubbles(data, config=config)
scatterplot.show()
```
